# XML

# Parsovanie XML dát
- XML (eXtensible Markup Language) je značkovací jazyk používaný na reprezentáciu hierarchických dát. 
- Na parsovanie XML v Pythone môžeme použiť knižnicu xml.etree.ElementTree.

```python
import xml.etree.ElementTree as ET

xml_data = '''<?xml version="1.0"?>
<data>
    <osoba>
        <meno>Adam</meno>
        <priezvisko>Šangala</priezvisko>
        <vek>33</vek>
        <mesto>Trnava</mesto>
    </osoba>
</data>'''

# root
koren = ET.fromstring(xml_data)

for osoba in koren.findall('osoba'):
    meno = osoba.find('meno').text
    vek = osoba.find('vek').text
    mesto = osoba.find('mesto').text
    print(f'Meno: {meno}, Vek: {vek}, Mesto: {mesto}')
```

# Načítanie XML súboru a vypísanie jeho obsahu

```python
import xml.etree.ElementTree as ET

# Načítanie XML súboru
tree = ET.parse('krajiny.xml')
root = tree.getroot()

# Vypísanie obsahu
for country in root.findall('country'):
    country_name = country.get('name')
    print(f'Krajina: {country_name}')
    for city in country.findall('city'):
        city_name = city.get('name')
        population = city.get('population')
        print(f' Mesto: {city_name}, Populácia: {population}')
```

# Parsovanie URL Adries
* Pri práci s webovými aplikáciami často potrebujeme analyzovať alebo konštruovať URL adresy. 
* Python poskytuje modul urllib na prácu s URL.

```python
from urllib.parse import urlparse, parse_qs

url = 'https://www.priklad.sk/index.php?id=123&category=hlavna'

parsovana_url = urlparse(url)
dopyt_parametre = parse_qs(parsovana_url.query)

print(f'Path: {parsovana_url.path},\nDopyt (Query) Parametre: {dopyt_parametre}')

url2 = 'https://www.vita.sk/?s=python'

parsovana_url2 = urlparse(url2)
dopyt_parametre2 = parse_qs(parsovana_url2.query)

print("----------------------\n")
print(f'Path: {parsovana_url2.path},\nDopyt (Query) Parametre: {dopyt_parametre2}')

url3 = 'https://www.vita.sk/online-kurzy-programovacie-jazyky/?orderby=&filter%5Bpa_uroven-obtiaznosti-kurzu%5D%5B%5D=253&filter%5Bpa_tema%5D%5B%5D=279'

parsovana_url3 = urlparse(url3)
dopyt_parametre3 = parse_qs(parsovana_url3.query)

print("----------------------\n")
print(f'Path: {parsovana_url3.path},\nDopyt (Query) Parametre: {dopyt_parametre3}')
```

# Parsovanie Log Súborov
* Log súbory obsahujú veľké množstvo dát generovaných aplikáciami, ktoré môžu byť užitočné pre analýzu a monitorovanie. 
* Tieto súbory často majú špecifický formát, ktorý je možné analyzovať regulárnymi výrazmi v Pythone.

```python
import re

log_data = """
INFO 2024-04-01 12:00:01 Vzorova Sluzba - Start procesu
ERROR 2024-04-01 12:01:35 Vzorova Sluzba - Chyba pri spracovaní údajov
ERROR 2024-04-01 12:01:48 Vzorova Sluzba - Chyba pri spracovaní údajov
INFO 2024-04-01 12:02:12 Vzorova Sluzba - Proces dokončený úspešne
"""

pattern = re.compile(r'ERROR (.*) - (.*)')
for riadok in log_data.split('\n'):
    zhoda = pattern.search(riadok)
    if zhoda:
        print(f'Čas: {zhoda.group(1)}, Správa: {zhoda.group(2)}')
```

# Parsovanie komplexných textových dát pomocou regulárnych výrazov
Regulárne výrazy sú mocným nástrojom na parsovanie a manipuláciu s textovými reťazcami, umožňujúci identifikovať vzory v texte.

```python
import re

text = "Email: miroslav.reiter@gmail.com, Phone: +421908123456"
email_pattern = r"[\w\.-]+@[\w\.-]+"
telefon_pattern = r"\+\d{12}"

email = re.search(email_pattern, text).group()
telefon = re.search(telefon_pattern, text).group()

print(f'Email: {email}, Telefón: {telefon}')
```

# Parsovanie Titulky (Title)

```python
from bs4 import BeautifulSoup

html_doc = """
<html><head><title>Moja stránka</title></head>
<body><p class="nadpis">Vitajte na mojej stránke!</p></body>
</html>
"""

soup = BeautifulSoup(html_doc, 'html.parser')
print(soup.title.text)  # Vypíše: Moja stránka
```

# Parsovanie Nadpisy (Headings)

```python
from bs4 import BeautifulSoup

# Simulované HTML dokumentu
html_doc = """
<html>
<head>
<title>Názov stránky</title>
</head>
<body>
<h1>Môj prvý nadpis</h1>
<p>Môj prvý odstavec.</p>
<h2>Druhý nadpis</h2>
<p>Druhý odstavec.</p>
</body>
</html>
"""

# Parsovanie HTML
soup = BeautifulSoup(html_doc, 'html.parser')

# Vyhľadávanie všetkých nadpisov h1 a h2
nadpisy = soup.find_all(['h1', 'h2'])

for nadpis in nadpisy:
    print(nadpis.text)
```

# Parsovanie Odkazov (Anchors)

```python
from bs4 import BeautifulSoup

# Simulované HTML s odkazmi
html_doc = """
<html>
<body>
<a href="https://google.sk/prvy" class="sibling">1. Odkaz</a>
<a href="https://google.sk/druhy" class="sibling">2. Odkaz</a>
</body>
</html>
"""

soup = BeautifulSoup(html_doc, 'html.parser')

# Vyhľadávanie všetkých odkazov
odkazy = soup.find_all('a')

for odkaz in odkazy:
    print(odkaz['href'])  # Vypíše URL odkazu
```

# Parsovanie Produktov (class, DIV a P)

```python
from bs4 import BeautifulSoup

# Simulované HTML
html_doc = """
<div class="product">
    <p class="name">Produkt 1</p>
    <p class="price">Cena: 100€</p>
</div>
<div class="product">
    <p class="name">Produkt 2</p>
    <p class="price">Cena: 200€</p>
</div>
"""

soup = BeautifulSoup(html_doc, 'html.parser')

# Vyhľadávanie pomocou CSS selektorov
products = soup.select('.product .name')

for product in products:
    print(product.text)  # Vypíše názvy produktov
```

# Zmena Obsahu a Sprehľadnenie (Prettify)

```python
from bs4 import BeautifulSoup

html_doc = """
<html>
<body>
<p id="uvod">Toto je uvodný odstavec.</p>
</body>
</html>
"""

soup = BeautifulSoup(html_doc, 'html.parser')

# Zmena textu v elemente
uvod = soup.find('p', id='uvod')
uvod.string = "Toto je zmenený uvodný odstavec."

print(soup.prettify())
```

# Jednoduché klastrovanie
- Použijeme algoritmus K-means na klastrovanie náhodne generovaných bodov do skupín. 
- Tento príklad využije scikit-learn, populárnu knižnicu pre strojové učenie.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Generujeme náhodné dáta
X = np.random.rand(100, 2)

# Použijeme K-means algoritmus na klastrovanie dát do 3 skupín
kmeans = KMeans(n_clusters=3, n_init=10)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Vizualizujeme klastre a centrá klastrov
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5)
plt.title("Vizualizácia výsledkov klastrovania")
plt.xlabel("X hodnota")
plt.ylabel("Y hodnota")
plt.show()
```

# Plotly (Dynamické Grafy)

```python
from sklearn.cluster import KMeans
import numpy as np
import plotly.graph_objects as go
import plotly.express as px

# Generujeme náhodné dáta
X = np.random.rand(100, 2)

# Použijeme K-means algoritmus na klastrovanie dát do 3 skupín
kmeans = KMeans(n_clusters=3, n_init=10)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Centrá klastrov
centers = kmeans.cluster_centers_

# Vytvoríme graf
fig = go.Figure()

# Pridáme body klastrov
for i in range(3):
    fig.add_trace(go.Scatter(x=X[y_kmeans == i, 0], y=X[y_kmeans == i, 1], mode='markers', name=f'Klaster {i+1}'))

# Pridáme centrá klastrov
fig.add_trace(go.Scatter(x=centers[:, 0], y=centers[:, 1], mode='markers', marker=dict(color='red', size=10, line=dict(color='DarkSlateGrey', width=2)), name='Centrá klastrov'))

# Aktualizujeme rozloženie pre lepšiu vizualizáciu
fig.update_layout(title='Vizualizácia výsledkov klastrovania s použitím Plotly', xaxis_title='X hodnota', yaxis_title='Y hodnota', legend_title='Legenda')

# Zobrazíme graf
fig.show()
```

# Pyparsing 

# Pyparsing
* umožňuje jednoduchú a intuitívnu analýzu a manipuláciu s textovými dátami bez nutnosti vytvárať komplikované gramatiky alebo sa zapodievať do detailov regulárnych výrazov. 
* Je obzvlášť užitočná pre analýzu dát a pre transformáciu dátových formátov. 

```python
from pyparsing import Word, nums, oneOf

# Definícia základných stavebných blokov
number = Word(nums)  # Čísla
operator = oneOf("+ - * /")  # Operátory

# Vytvorenie gramatiky
expression = number + operator + number

# Analyzovanie výrazu
result = expression.parseString("3 + 4")

print(result)  # Výstup: ['3', '+', '4']
```

# Analýza Logického Výrazu
Ak by sme chceli analyzovať trochu zložitejší logický výraz obsahujúci zátvorky a logické operátory

```python
from pyparsing import Word, alphas, opAssoc, infixNotation

# Definícia základných prvkov
variable = Word(alphas)
and_ = "a"
or_ = "alebo"
not_ = "nie"

# Definícia gramatiky pre logický výraz
expr = infixNotation(variable,
    [
        (not_, 1, opAssoc.RIGHT),
        (and_, 2, opAssoc.LEFT),
        (or_, 2, opAssoc.LEFT),
    ])

# Analyzovanie výrazu
result = expr.parseString("A a nie B alebo C")

print(result.asList())  # Výstup ukáže vnorenú štruktúru výrazu
print(result)  # Výstup 
print(expr)  # Výstup 
```

# Analýza a Transformácia Dát
Pyparsing je tiež skvelý na transformáciu textových dát. Predstavme si, že chceme transformovať dátový záznam zo zložitého formátu na slovník

```python
from pyparsing import Word, alphas, alphanums, Suppress, Group, Dict

# Elementy gramatiky
identifier = Word(alphas, alphanums + "_").setName("identifier")
equals = Suppress("=")
value = Word(alphanums).setName("value")

# Pravidlo pre záznam
record = Group(identifier + equals + value)

# Definícia celého vstupu
config = Dict(record[...])

# Analýza a transformácia dát
data = "meno=Adam vek=33"
result = config.parseString(data)

print(dict(result))  # Výstup: {'meno': 'Adam', 'vek': '33'}
```

```python
from pyparsing import Word, alphanums, Combine, Literal

# Definícia častí emailovej adresy
localPart = Word(alphanums + "._")
atSymbol = Literal("@")
domain = Word(alphanums + ".")
emailAddress = Combine(localPart + atSymbol + domain)

# Analýza a transformácia
email = "miroslav.reiter@gmail.com"
transformed = emailAddress.transformString("Osobný email: miroslav.reiter@gmail.com")
transformed2 = emailAddress.transformString("Zastihnete ma na firemnom emaily riaditel@itacademy.sk alebo v Bratislave")

print(transformed)  # Výstup: 'Contact: example.user@domain.com'
print(transformed2)  # Výstup: 'Contact: example.user@domain.com'
```

```python
from pyparsing import oneOf, OneOrMore, Word, alphas

# Preddefinované klúčové slová
keywords = oneOf("Python Pyparsing analýza dáta transformácia")

# Pravidlo pre slová, ktoré nie sú klúčové slová
nonKeyword = Word(alphas)

# Gramatika, ktorá ignoruje všetko okrem klúčových slov
grammar = OneOrMore(keywords | nonKeyword.suppress())

# Analyzovanie textu
text = "Pyparsing je knižnica v Python pre analýza a transformácia dáta."
result = grammar.searchString(text)

print(result)  # Výstup zobrazí len klúčové slová
```

```python
from pyparsing import makeHTMLTags, SkipTo

# Vytvorenie pravidiel pre HTML link tag
anchorStart, anchorEnd = makeHTMLTags("a")
link = anchorStart + SkipTo(anchorEnd).setResultsName("url") + anchorEnd

# Príklad HTML textu
htmlText = '<a href="https://google.sk">Stranka Google.sk</a>'

# Vyhľadávanie a extrakcia linku
for token, start, end in link.scanString(htmlText):
    print(token.url)  # Výstup: 'Stranka Google.sk'
```
